```python
import pandas as pd

insurance_1_df = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv")
insurance_2_df = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv")
insurance_3_df = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv")
```

```python
# Estandarizamos las columnas de los 3 dataframes, dejandolas en el orden del primer dataframe

# Primero hacer un copy del data frame que queremos modificar, en este caso insurance_2_df
# Segundo especificamos las columnas en el orden que las queremos: 'Customer', 'ST', 'GENDER', 'Education', 'Customer Lifetime Value',
#'Income', 'Monthly Premium Auto', 'Number of Open Complaints','Policy Type', 'Vehicle Class', 'Total Claim Amount'

insurance_2_df_copy = insurance_2_df.copy()

insurance_2_df_copy = insurance_2_df_copy[[
    'Customer', 
    'ST',
    'GENDER', 
    'Education', 
    'Customer Lifetime Value',
    'Income', 
    'Monthly Premium Auto', 
    'Number of Open Complaints',
    'Policy Type', 
    'Vehicle Class', 
    'Total Claim Amount']]

insurance_2_df_copy.head()
```

```python
insurance_3_df.columns
```

```python
# En el caso del dataframe 3, de manera previa a reorganizar las columnas debemos cambiar el 'ST' por 'State'

insurance_3_df_copy = insurance_3_df.copy()
insurance_3_df_copy
```

```python
insurance_1_df.columns
```

```python
insurance_3_df_copy.columns
```

```python
insurance_3_df_copy= insurance_3_df_copy.rename(columns={"State":"ST", "Gender":"GENDER"})
insurance_3_df_copy
```

```python
insurance_3_df_copy = insurance_3_df_copy[[
    'Customer', 
    'ST',
    'GENDER', 
    'Education', 
    'Customer Lifetime Value',
    'Income', 
    'Monthly Premium Auto', 
    'Number of Open Complaints',
    'Policy Type', 
    'Vehicle Class', 
    'Total Claim Amount']]

insurance_3_df_copy.head()
```

```python
insurance_3_df_copy= insurance_3_df_copy.rename(columns={"Gender":"GENDER"})
insurance_3_df_copy
```

```python
insurance_1_df.columns
```

```python
insurance_2_df_copy.columns
```

```python
insurance_3_df_copy.columns
```

```python
insurance_completa_df=pd.concat([insurance_1_df,insurance_2_df_copy,insurance_3_df_copy])
insurance_completa_df.info()
```

```python
insurance_completa_df_1 = insurance_completa_df.copy()
insurance_completa_df_1.columns = [column.lower().replace(" ", "_") for column in insurance_completa_df_1.columns]
insurance_completa_df_1.info()
```

```python
insurance_completa_df_1.nunique()
```

```python
insurance_completa_df_1["st"].unique()
```

```python
state_mapping = {'Washington': 'Washington', 
                  'Arizona': 'Arizona',
                  'Nevada': 'Nevada',
                  'California':'California',
                  'Oregon': 'Oregon',
                  'Cali':'California',
                  'AZ':'Arizona',
                  'WA':'Washington',
                  'nan':'nan'}
insurance_completa_df_1['st'] = insurance_completa_df_1['st'].map(state_mapping)
insurance_completa_df_1.head()
```

```python
insurance_completa_df_1["st"].unique()
```

```python
insurance_completa_df_1["gender"].unique()
```

```python
gender_mapping = {'F': 'F', 
                  'Femal': 'F',
                  'M': 'M',
                  'Male':'M',
                  'female': 'F',
                  'nan':'nan'}
insurance_completa_df_1['gender'] = insurance_completa_df_1['gender'].map(gender_mapping)
insurance_completa_df_1.head()
```

```python
insurance_completa_df_1["gender"].unique()
```

```python
insurance_completa_df_1.info()
```

```python
insurance_completa_df_1["education"].unique()
```

```python
education_mapping = {'Master': 'Master', 
                  'Bachelor': 'Bachelor',
                  'High School or Below': 'High School or Below',
                  'Bachelors':'Bachelor',
                    'College': 'College',
                  'Doctor': 'Doctor',
                  'nan':'nan'}
insurance_completa_df_1['education'] = insurance_completa_df_1['education'].map(education_mapping)
insurance_completa_df_1.head()
```

```python
insurance_completa_df_1.info()
```

```python
insurance_completa_df_1["education"].unique()
```

```python
insurance_completa_df_1["education"].info()
```

```python
insurance_completa_df_1["policy_type"].unique()
```

```python
insurance_completa_df_1["vehicle_class"].unique()
```

```python
vehicle_class_mapping = {'Four-Door Car': 'Four-Door Car', 
                  'Two-Door Car': 'Two-Door Car',
                  'SUV': 'SUV',
                  'Luxury SUV':'Luxury',
                  'Sports Car': 'Luxury',
                  'nan':'nan',
                         'Luxury Car':'Luxury'}
insurance_completa_df_1['vehicle_class'] = insurance_completa_df_1['vehicle_class'].map(vehicle_class_mapping)
insurance_completa_df_1.head()
```

```python
insurance_completa_df_1["vehicle_class"].unique()
```

```python
insurance_completa_df_1['customer_lifetime_value'].info()
```

```python
insurance_completa_df_2 = insurance_completa_df_1.copy()
insurance_completa_df_2['customer_lifetime_value'] = insurance_completa_df_2['customer_lifetime_value'].apply(lambda x: x.replace('%', '') if isinstance(x, str) and x.endswith('%') else x)
insurance_completa_df_2
```

```python
insurance_completa_df_1.info()
```

```python
insurance_completa_df_2.info()
```

```python
insurance_completa_df_2['customer_lifetime_value'].info()
```

```python
insurance_completa_df_2['number_of_open_complaints'] = insurance_completa_df_2['number_of_open_complaints'].apply(lambda x: x.split('/')[1] if isinstance(x, str) else x)
```

```python
insurance_completa_df_2
```

```python
insurance_completa_df_2.info()
```

```python
insurance_completa_df_2["customer_lifetime_value"] = insurance_completa_df_2["customer_lifetime_value"].astype(float)
```

```python
insurance_completa_df_2["number_of_open_complaints"] = insurance_completa_df_2["number_of_open_complaints"].astype(float)
```

```python
insurance_completa_df_2.info()
```

```python
# Comprobamos la cantidad de valores nulos que existen en nuestro dataframe
insurance_completa_df_3 = insurance_completa_df_2.copy()
insurance_completa_df_3.isnull().sum()
```

```python
# Borrar todas las filas en las que todos los valores sean Nan
insurance_completa_df_3 = insurance_completa_df_3.dropna(how='all')
insurance_completa_df_3
```

```python
insurance_completa_df_3.isnull().sum()
```

```python
round(insurance_completa_df_3["gender"].isna().sum() / len(insurance_completa_df_3) * 100, 1)
```

```python
round(insurance_completa_df_3["education"].isna().sum() / len(insurance_completa_df_3) * 100, 1)
```

```python
round(insurance_completa_df_3["customer_lifetime_value"].isna().sum() / len(insurance_completa_df_3) * 100, 1)
```

```python
insurance_completa_df_3["gender"].bfill(inplace=True)
insurance_completa_df_3
```

```python
insurance_completa_df_3.isna().sum()
```

```python
mean_customer=insurance_completa_df_3["customer_lifetime_value"].mean()
mean_customer
```

```python
insurance_completa_df_3["customer_lifetime_value"]=insurance_completa_df_3["customer_lifetime_value"].fillna(mean_customer)
insurance_completa_df_3
```

```python
insurance_completa_df_3.isna().sum()
```

```python
# DETECTAR DUPLICADOS
insurance_completa_df_4 = insurance_completa_df_3.copy()
insurance_completa_df_4.duplicated().sum()
```

```python
# Como hay 3 duplicados, elegimos la estrategia de eliminar las 3 filas que estén duplicadas mediante el método drop_duplicates()
insurance_completa_df_5 = insurance_completa_df_4.drop_duplicates()
```

```python
# Comprobamos que hemos eliminado los duplicados:
insurance_completa_df_5.duplicated().sum()
```
